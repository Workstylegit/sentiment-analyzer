```python
# Install required tools
!pip install -q streamlit textblob
!npm install -g localtunnel

# Write your Wear Style app
%%writefile wearstyle_app.py
import streamlit as st
from textblob import TextBlob

# Analysis function
def analyze_review(text):
    score = round(TextBlob(str(text)).sentiment.polarity, 2)
    sentiment = "Positive 😊" if score > 0 else "Negative 😞" if score < 0 else "Neutral 😐"
    t = str(text).lower()
    topics = []
    if any(w in t for w in ["size", "fit", "small", "large", "tight", "loose"]):
        topics.append("Size & Fit")
    if any(w in t for w in ["fabric", "material", "soft", "rough", "thin"]):
        topics.append("Fabric Quality")
    if any(w in t for w in ["color", "fade", "shade"]):
        topics.append("Color & Look")
    if any(w in t for w in ["price", "cost", "worth"]):
        topics.append("Price")
    return sentiment, score, ", ".join(topics) if topics else "General"

# App interface
st.set_page_config(page_title="Wear Style Analyzer", layout="centered")
st.title("👕 Wear Style — Review Analyzer")
st.write("Paste any clothing review to check sentiment and topic.")

user_review = st.text_area("Enter your review here:", height=150)

if st.button("Analyze"):
    if user_review.strip():
        res, scr, top = analyze_review(user_review)
        st.success(f"**Result:** {res}")
        st.info(f"**Score:** {scr} | **Topic:** {top}")
    else:
        st.warning("Please type or paste a review first!")
```
